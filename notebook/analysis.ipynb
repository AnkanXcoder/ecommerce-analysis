```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/Online_Retail.csv')
df.sample(5)
```

```python
# 1. Load dataset and inspect (head, info, shape)

df.head()
df.sample()
df.shape
```

```python
# 2. Check missing values column-wise
df.isnull().sum()
```

```python
# 3. Convert InvoiceDate to datetime

df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df.info()
```

```python
# 4. Remove rows where CustomerID is null

df = df.dropna(subset=['CustomerID'])
df.shape
```

```python
# 5. Filter only positive Quantity and UnitPrice

df = df[(df['Quantity'] > 0) & (df['UnitPrice'] > 0)]
df.head()
```

```python
# 6. Create new column: TotalPrice = Quantity * UnitPrice

df['TotalPrice'] = df['Quantity']*df['UnitPrice']
df.head()
```

```python
# 7. Find number of unique customers

df['CustomerID'].nunique()
```

```python
# 8. Find number of unique products

df['StockCode'].nunique()
```

```python
# 9. Transactions from United Kingdom only

uk_df = df[df['Country'] == 'United Kingdom']
uk_df
```

```python
# 10. Top 10 most expensive transactions
top_10_most_expensive = df.sort_values('TotalPrice',ascending=False).head(10)
top_10_most_expensive
```

```python
# 11. Find all cancelled invoices (InvoiceNo starts with 'C')

df[df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
# 12. Remove cancelled transactions
df = df[~df['InvoiceNo'].astype(str).str.startswith('C')]
```

```python
# 13. Sort customers by total spending
df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False)
```

```python
# 14. Total revenue per country
df.groupby('Country')['TotalPrice'].sum().sort_values(ascending=False)
```

```python
# 15. Total orders per customer

df.groupby('CustomerID')['InvoiceNo'].nunique().sort_values(ascending=False)
```

```python
# 16. Average order value per customer

(df.groupby('CustomerID')['TotalPrice'].sum()/df.groupby('CustomerID')['InvoiceNo'].nunique()).sort_values(ascending=False)
```

```python
# 17. Top 10 customers by revenue

df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False).head(10)
```

```python
# 18. Most sold products (by quantity)

df.groupby('StockCode')['Quantity'].sum().sort_values(ascending=False)
```

```python
# 19. Revenue per product

df.groupby('StockCode')['TotalPrice'].sum().sort_values(ascending=False)
```

```python
# 20. Revenue by (Country, CustomerID)

df.groupby(['Country','CustomerID'])['TotalPrice'].sum().sort_values(ascending=False)
```

```python
# 21. Quantity sold by (Country, StockCode)

df.groupby(['Country','StockCode'])['Quantity'].sum().sort_values(ascending=False)
```

```python
# 22. Reset index after groupby

df.groupby(['Country','CustomerID'])['TotalPrice'].sum().sort_values(ascending=False).reset_index()
```

```python
# 23. Sort multi-index results

df.groupby(['Country','CustomerID'])['TotalPrice'].sum().reset_index().sort_values(['Country','TotalPrice'],ascending=[True,False])
```

```python
# 24. Extract year, month, day from InvoiceDate

df['year'] = df['InvoiceDate'].dt.year
df['month'] = df['InvoiceDate'].dt.month
df['day'] = df['InvoiceDate'].dt.day
df.head()
```

```python
df['month'] = df['InvoiceDate'].dt.month
```

```python
# 25. Monthly revenue

df.groupby('month')['TotalPrice'].sum()
```

```python
# 26. Daily revenue trend

df['date'] = df['InvoiceDate'].dt.date
df.groupby('date')['TotalPrice'].sum()
```

```python
# 27. Find busiest hour of the day

df['hour'] = df['InvoiceDate'].dt.hour
df.groupby('hour')['InvoiceNo'].nunique().sort_values(ascending=False).reset_index(name='order_count').head(1)
```

```python
# 28. Revenue by weekday
df.groupby(df['InvoiceDate'].dt.day_name())['TotalPrice'].sum().sort_values(ascending=False)
```

```python
df['day_num'] = df['InvoiceDate'].dt.dayofweek
df['day'] = df['InvoiceDate'].dt.day_name()
df.groupby(['day_num','day'])['TotalPrice'].sum().sort_index()
```

```python
# 29. Number of orders per month

df.groupby('month')['InvoiceNo'].nunique()
```

```python
# 30. Pivot: Revenue per Country vs Month

df['month_name'] = df['InvoiceDate'].dt.month_name()
df.pivot_table(index='Country',columns='month_name',values='TotalPrice',aggfunc='sum')
```

```python
# 31. Pivot: Quantity per Product vs Country

df.pivot_table(index='StockCode',columns='Country',values='Quantity',aggfunc='sum')
```

```python
# 32. Fill missing values in pivot

df.pivot_table(index='StockCode',columns='Country',values='Quantity',aggfunc='sum',fill_value=0)
```

```python
# 33. Add margins (totals)

df.pivot_table(index='StockCode',columns='Country',values='Quantity',aggfunc='sum',fill_value=0,margins=True,margins_name='Total')
```

```python
# 34. Clean Description (lowercase, strip spaces)

df['Description'] = df['Description'].str.lower().str.strip()
```

```python
# 35. Find products containing word "set"
df[df['Description'].str.contains('set',case=False,na=False)]
```

```python
# 36. Count most frequent words in descriptions

df['words'] = df['Description'].str.split(' ')

df.explode('words')['words'].value_counts().sort_values(ascending=False)
```

```python
# 37. Remove special characters

df['Description'] = df['Description'].str.replace(r'[^a-zA-Z0-9\s]', '', regex=True)
```

```python
# 38. Split dataset into:
# Customers table
# Transactions table

cus_table = df[['CustomerID','Country']].drop_duplicates()
tran_table = df[['InvoiceNo','StockCode','Description','CustomerID','Quantity','InvoiceDate','UnitPrice','TotalPrice']]
```

```python
# 39. Merge them back

pd.merge(tran_table,cus_table,how='inner',on='CustomerID')
```

```python
# 40. Inner join vs left join comparison
inner_shape = pd.merge(tran_table,cus_table,how='inner',on='CustomerID').shape[0]
left_shape = pd.merge(tran_table,cus_table,how='left',on='CustomerID').shape[0]
print(f"inner join row {inner_shape}")
print(f"left join row {left_shape}")

if left_shape == inner_shape:
  print("no missing customers")
else:
  print("some transactions have no matching customer")
```

```python
# 41. Find customers with no transactions

merged = pd.merge(cus_table,tran_table,how='left',on='CustomerID')

merged[merged['InvoiceNo'].isnull()]
```

```python
# 42. Top country by revenue

df.groupby('Country')['TotalPrice'].sum().idxmax()
```

```python
# 43. Repeat vs new customers

df['repeat'] = df.groupby('CustomerID')['InvoiceNo'].transform('nunique')

df['customer_type'] = 'New customer'
df.loc[df['repeat'] > 1, 'customer_type'] = 'Repeat customer'
```

```python
# 44. Customer lifetime value (CLV)

df.groupby('CustomerID')['TotalPrice'].sum().sort_values(ascending=False)
```

```python
# 45. Most profitable month

df.groupby('month')['TotalPrice'].sum().idxmax()
```

```python
# 46. Cohort analysis

df['first_month'] = df.groupby('CustomerID')['InvoiceDate'].transform('min')
df['order_month'] = df['InvoiceDate'].dt.to_period('M')
df['first_month'] = df['first_month'].dt.to_period('M')

df.groupby(['first_month','order_month'])['CustomerID'].nunique()
```

```python
# 47. Find top 3 products per country
df.groupby(['Country','StockCode'])['TotalPrice'].sum().groupby(level=0).nlargest(3)
```

```python
# 48. Find customers who purchased more than 5 times

# df[df.groupby('CustomerID')['InvoiceNo'].transform('nunique') > 5]['CustomerID']

df.groupby('CustomerID')['InvoiceNo'].nunique().loc[lambda x: x > 5].index
```

```python
# 49. Detect outliers in spending

spending = df.groupby('CustomerID')['TotalPrice'].sum()
threshold = spending.quantile(0.95)
outliers = spending[spending > threshold]
outliers
```

```python
# 50. Rank customers within each country'

df['cus_con'] = df.groupby(['Country','CustomerID'])['TotalPrice'].transform('sum')

df['rank_of_the_customer'] = df.groupby('Country')['cus_con'].rank(ascending=False, method='dense')
```

```python
# 51. Rolling monthly revenue

df.groupby('date')['TotalPrice'].sum().sort_index().cumsum()
```

```python
# 52. Cumulative revenue over time

df.groupby(df['InvoiceDate'].dt.date)['TotalPrice'].sum().sort_index().cumsum()
```

```python
# 53. Build full report:
#     Monthly revenue
#     Top customers
#     Top products
#     Country-wise analysis


report = df.groupby('Country').agg({
    'TotalPrice': 'sum',
    'InvoiceNo': 'nunique',
    'CustomerID': 'nunique'
}).rename(columns={
    'TotalPrice': 'Total_Revenue',
    'InvoiceNo': 'Total_Orders',
    'CustomerID': 'Unique_Customers'
})

report['AOV'] = report['Total_Revenue'] / report['Total_Orders']

report = report.sort_values('Total_Revenue', ascending=False)

report
```
